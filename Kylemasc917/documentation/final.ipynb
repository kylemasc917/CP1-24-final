```python
import os
from final import extract_temp_from_markdown  # Import your function

def process_markdown_files(folder_path):
    """Process all markdown files in a given folder."""
    temperatures_dict = {}
    
    # List all markdown files in the folder
    markdown_files = [f for f in os.listdir(folder_path) if f.endswith('sinewalk.md')]

    # Iterate over all markdown files and extract temperatures
    for file_name in markdown_files:
        file_path = os.path.join(folder_path, file_name)
        temperatures = extract_temp_from_markdown(file_path)
        temperatures_dict[file_name] = temperatures

    return temperatures_dict

# Specify the folder containing your markdown files
folder_path = '/workspaces/CP1-24-final/Kylemasc917/data'  # Replace with your folder path

# Process the files
temperatures_dict = process_markdown_files(folder_path)

# Display the extracted temperatures
for file_name, temps in temperatures_dict.items():
    print(f"Temperatures from {file_name}: {temps}")
```
